```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_loan = pd.read_csv('prosperLoanData.csv')
```

```python
def labels(title,x,y):
    return plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    
def fig_size(a,b):
    return plt.figure(figsize=(a,b))
```

## (Visualization 1) - Borrowers state

>For the first visualization, we would look at the borrowers state category. The aim is to get the states the borrowers came from and what state tops the chart?

>For this visualization, California has the highest number, double the percentage compared to any other state. followed by Texas, New York, and Florida follows suit.

```python
fig_size(12,7);
color = sns.color_palette()[0]
order = df_loan.BorrowerState.value_counts().iloc[:10].index
sns.countplot(data=df_loan, y='BorrowerState', color=color, order=order)
labels('Borrower State','State Abbreviation','Frequency');
```

## (Visualization 2) - Relationship between Loan Amount and Income Range

> For this Visualization, Using a boxplot we will visualize if the IncomeRange of the borrowers affect the LoanOriginalAmount.

>It can be observed that there is a little correlation between income range and the amount given as loan. Judging from the boxplot the higher income range the higher the loan amount. The highest loan amount belongs to people who earn 100k+, and the lowest loan amount to people who don't earn anything.

```python
color = sns.color_palette()[0]
fig_size(12,7);
order = ['Not displayed','Notemployed','$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999','$100,000+']
sns.boxplot(data=df_loan, x='IncomeRange', y='LoanOriginalAmount', color=color, order=order)
#plt.xticks(rotation=15)
plt.title("Relationship between Loan Amount and Income Range")
plt.xlabel("IncomeRange")
plt.ylabel("Original Amount");
```

## (Visualization 3) - Employment Status

>Finally, this visualization will give you a pictorial representation of the category of people in variable 'EmploymentStatus'. we get to see people that took the highest loan.

>And we can see that the borrowers that are Employed rank the highest. And it is understandable why part-time, Not-employed and Retired took the least loan.

```python
fig_size(10,7)
color = sns.color_palette()[0]
order = df_loan.EmploymentStatus.value_counts().index
sns.countplot(data=df_loan, x='EmploymentStatus', color=color, order=order)
plt.title("Employment Status")
plt.xlabel("Status")
plt.ylabel("Frequency");
```
